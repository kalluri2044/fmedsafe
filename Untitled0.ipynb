```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset
np.random.seed(0)
data = {
    'Category': np.random.choice(['A', 'B', 'C', 'D'], size=100),
    'Value': np.random.randint(1, 100, size=100),
    'X': np.random.rand(100),
    'Y': np.random.rand(100) * 100
}

df = pd.DataFrame(data)

# Set the style for seaborn
sns.set(style="whitegrid")

# 1. Bar Chart
plt.figure(figsize=(10, 5))
sns.countplot(x='Category', data=df)
plt.title('Bar Chart of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# 2. Pie Chart
plt.figure(figsize=(8, 8))
df['Category'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Categories')
plt.ylabel('')  # Hide the y-label
plt.show()

# 3. Scatter Plot
plt.figure(figsize=(10, 5))
plt.scatter(df['X'], df['Y'], alpha=0.6)
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# 4. Line Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=df.groupby('Category')['Value'].mean().reset_index(), x='Category', y='Value', marker='o')
plt.title('Line Plot of Average Value by Category')
plt.xlabel('Category')
plt.ylabel('Average Value')
plt.show()

# 5. Strip Plot
plt.figure(figsize=(10, 5))
sns.stripplot(x='Category', y='Value', data=df, jitter=True)
plt.title('Strip Plot of Value by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# 6. Swarm Plot
plt.figure(figsize=(10, 5))
sns.swarmplot(x='Category', y='Value', data=df)
plt.title('Swarm Plot of Value by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# 7. Histogram
plt.figure(figsize=(10, 5))
sns.histplot(df['Value'], bins=15, kde=False)
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 8. Density Plot
plt.figure(figsize=(10, 5))
sns.kdeplot(df['Value'], fill=True)
plt.title('Density Plot of Values')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Generate Sample Data
np.random.seed(42)
categories = ['A', 'B', 'C', 'D']
values = [40, 70, 30, 90]
x = np.random.rand(50) * 10
y = np.random.rand(50) * 100
data = np.random.normal(50, 15, 200)
category_labels = np.random.choice(['A', 'B', 'C'], size=30)
data_strip = np.random.randint(10, 100, size=30)

# Create a DataFrame for complex plots
df = pd.DataFrame({'Category': np.random.choice(categories, 100), 'Value': np.random.randn(100) * 10 + 50})

plt.figure(figsize=(18, 16))

# 1. Bar Chart
plt.subplot(4, 4, 1)
plt.bar(categories, values, color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart')

# 2. Pie Chart
plt.subplot(4, 4, 2)
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=['blue', 'orange', 'green', 'red'])
plt.title('Pie Chart')

# 3. Scatter Plot
plt.subplot(4, 4, 3)
plt.scatter(x, y, color='purple', alpha=0.6)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot')

# 4. Line Plot
plt.subplot(4, 4, 4)
plt.plot(range(1, 11), np.random.randint(10, 50, size=10), marker='o', linestyle='-', color='blue')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot')

# 5. Strip Plot
plt.subplot(4, 4, 5)
sns.stripplot(x=category_labels, y=data_strip, jitter=True, palette='Set1')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Strip Plot')

# 6. Swarm Plot
plt.subplot(4, 4, 6)
sns.swarmplot(x=category_labels, y=data_strip, palette='Set2')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Swarm Plot')

# 7. Histogram
plt.subplot(4, 4, 7)
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# 8. Density Plot (KDE)
plt.subplot(4, 4, 8)
sns.kdeplot(data, shade=True, color='red')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Plot')

# 9. Rug Plot
plt.subplot(4, 4, 9)
sns.rugplot(data, color='black')
plt.xlabel('Value')
plt.title('Rug Plot')

# 10. Box Plot
plt.subplot(4, 4, 10)
sns.boxplot(y=data, color='lightblue')
plt.title('Box Plot')

# 11. Violin Plot
plt.subplot(4, 4, 11)
sns.violinplot(y=data, color='purple')
plt.title('Violin Plot')

# 12. Ridgeline Plot
plt.subplot(4, 4, 12)
sns.kdeplot(df[df["Category"] == 'A']["Value"], shade=True, label='A')
sns.kdeplot(df[df["Category"] == 'B']["Value"], shade=True, label='B')
sns.kdeplot(df[df["Category"] == 'C']["Value"], shade=True, label='C')
sns.kdeplot(df[df["Category"] == 'D']["Value"], shade=True, label='D')
plt.legend()
plt.title('Ridgeline Plot')

# 13. Bee Swarm Plot
plt.subplot(4, 4, 13)
sns.swarmplot(x="Category", y="Value", data=df, palette='Set1')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bee Swarm Plot')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic

# Generate Sample Data
np.random.seed(42)
df = pd.DataFrame({
    'Category1': np.random.choice(['A', 'B', 'C'], 100),
    'Category2': np.random.choice(['X', 'Y'], 100),
    'Continuous1': np.random.randn(100) * 10 + 50,
    'Continuous2': np.random.randn(100) * 5 + 30
})

plt.figure(figsize=(18, 20))

# 1. Stacked Bar Chart (Categorical vs. Categorical)
plt.subplot(4, 3, 1)
df.groupby(['Category1', 'Category2']).size().unstack().plot(kind='bar', stacked=True, colormap='coolwarm', ax=plt.gca())
plt.title("Stacked Bar Chart")

# 2. Grouped Bar Chart (Categorical vs. Categorical)
plt.subplot(4, 3, 2)
df.groupby(['Category1', 'Category2']).size().unstack().plot(kind='bar', colormap='viridis', ax=plt.gca())
plt.title("Grouped Bar Chart")

# 3. Segmented Bar Chart (Categorical vs. Categorical)
plt.subplot(4, 3, 3)
df.groupby(['Category1', 'Category2']).size().unstack().apply(lambda x: x / x.sum(), axis=1).plot(kind='bar', stacked=True, colormap='Set2', ax=plt.gca())
plt.title("Segmented Bar Chart")

# 4. Mosaic Plot (Categorical vs. Categorical)
plt.subplot(4, 3, 4)
mosaic(df, ['Category1', 'Category2'], ax=plt.gca(), title="Mosaic Plot")

# 5. Scatter Plot with Fit Line (Continuous vs. Continuous)
plt.subplot(4, 3, 5)
sns.regplot(x='Continuous1', y='Continuous2', data=df, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'}, ax=plt.gca())
plt.title("Scatter Plot with Fit Line")

# 6. Bar Chart (Summary Statistics, Categorical vs. Continuous)
plt.subplot(4, 3, 6)
sns.barplot(x='Category1', y='Continuous1', data=df, estimator=np.mean, ci='sd', palette='Set1', ax=plt.gca())
plt.title("Bar Chart (Summary Statistics)")

# 7. Grouped Kernel Density Plot (Categorical vs. Continuous)
plt.subplot(4, 3, 7)
sns.kdeplot(data=df, x='Continuous1', hue='Category1', fill=True, common_norm=False, palette='coolwarm', ax=plt.gca())
plt.title("Grouped Kernel Density Plot")

# 8. Box Plot (Categorical vs. Continuous)
plt.subplot(4, 3, 8)
sns.boxplot(x='Category1', y='Continuous1', data=df, palette='pastel', ax=plt.gca())
plt.title("Box Plot")

# 9. Violin Plot (Categorical vs. Continuous)
plt.subplot(4, 3, 9)
sns.violinplot(x='Category1', y='Continuous1', data=df, palette='muted', ax=plt.gca())
plt.title("Violin Plot")

# 10. Ridgeline Plot (Categorical vs. Continuous)
plt.subplot(4, 3, 10)
sns.kdeplot(df[df["Category1"] == 'A']["Continuous1"], shade=True, label='A')
sns.kdeplot(df[df["Category1"] == 'B']["Continuous1"], shade=True, label='B')
sns.kdeplot(df[df["Category1"] == 'C']["Continuous1"], shade=True, label='C')
plt.legend()
plt.title("Ridgeline Plot")

# 11. Beeswarm Plot (Categorical vs. Continuous)
plt.subplot(4, 3, 11)
sns.swarmplot(x="Category1", y="Continuous1", data=df, palette='Set1', ax=plt.gca())
plt.title("Beeswarm Plot")

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic

# Sample Stock Market Data
data = {
    "Stock Index": ["DFM", "ADSM", "BA", "KSE", "MSM30", "DSM20", "SE"],
    "Daily Volatility (Normal)": [1.81, 1.32, 0.58, 0.71, 0.79, 1.48, 1.86],
    "Daily Volatility (Severe)": [12.2, 7.1, 3.8, 3.7, 8.7, 8.1, 11.0],
    "Annual Volatility (Normal)": [29.2, 21.4, 9.4, 11.5, 12.6, 23.9, 30.0],
    "Annual Volatility (Severe)": [196.0, 114.1, 60.8, 52.6, 140.3, 130.2, 177.9],
    "Sensitivity Factor": [0.58, 0.40, 0.06, 0.12, 0.10, 0.31, 0.98]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Stacked Bar Chart
plt.figure(figsize=(12, 6))
df.set_index("Stock Index")[["Daily Volatility (Normal)", "Daily Volatility (Severe)"]].plot(kind="bar", stacked=True, colormap="coolwarm", alpha=0.8)
plt.title("Stacked Bar Chart - Daily Volatility")
plt.ylabel("Volatility (%)")
plt.legend(title="Market Condition")
plt.xticks(rotation=45)
plt.show()

# Grouped Bar Chart
plt.figure(figsize=(12, 6))
df.set_index("Stock Index")[["Annual Volatility (Normal)", "Annual Volatility (Severe)"]].plot(kind="bar", colormap="viridis", alpha=0.8)
plt.title("Grouped Bar Chart - Annual Volatility")
plt.ylabel("Volatility (%)")
plt.legend(title="Market Condition")
plt.xticks(rotation=45)
plt.show()

# Segmented Bar Chart (Relative Proportions)
df["Total Annual Volatility"] = df["Annual Volatility (Normal)"] + df["Annual Volatility (Severe)"]
df["Normal %"] = df["Annual Volatility (Normal)"] / df["Total Annual Volatility"]
df["Severe %"] = df["Annual Volatility (Severe)"] / df["Total Annual Volatility"]

df[["Stock Index", "Normal %", "Severe %"]].set_index("Stock Index").plot(kind="bar", stacked=True, colormap="coolwarm", alpha=0.8)
plt.title("Segmented Bar Chart - Annual Volatility %")
plt.ylabel("Proportion")
plt.xticks(rotation=45)
plt.show()

# Mosaic Plot
mosaic(df, ["Stock Index", "Sensitivity Factor"], title="Mosaic Plot - Sensitivity Factor by Stock Index", label_rotation=90)
plt.show()
```
